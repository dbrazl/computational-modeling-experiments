# Runge Kutta Ordem 2

 um mtodo para resoluo de Equaes Diferenciais Ordinarias (EDO) de primeira ordem.

EDO que iremos resolver:



Intervalo: [0,1]

 um problema de valor inicial.

---

Reviso de Runge Kutta 2:



```python
import numpy as np
```

```python
def edo(x,y):
  return y*np.cos(x)
```

```python
x0 = 0
y0 = 1
```

```python
a = 0
b = 1
N = 10
```

```python
h = (b - a) / N
```

```python
xi = [x0]
yi = [y0]
```

```python
for index in range(N):
  k1 = edo(x0, y0)
  k2 = edo(x0 + h, y0 + h * k1)
  yk = y0 + 0.5 * (k1 + k2) * h
  x0 = x0 + h
  y0 = yk

  xi.append(x0)
  yi.append(yk)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(xi, yi)
```

```python
plt.plot(xi, yi, 'o')
```
