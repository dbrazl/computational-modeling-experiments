# Estresse concentrado em um pequeno buraco circular em um plato sobre tens√£o cisalhante pura


```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 100)

y1 = np.sin(2 * theta)
y2 = np.cos(2 * theta)
y3 = np.sin(3 * theta)

plt.figure(figsize=(10, 5))
plt.plot(theta, y1, label=r'$\sin(2\theta)$')
plt.plot(theta, y2, label=r'$\cos(2\theta)$')
plt.plot(theta, y3, label=r'$\sin(2\theta)$', linestyle='--')

plt.xlabel(r'$\theta$ (radianos)')
plt.ylabel('Amplitude')
plt.title(r'Gr√°ficos de $\sin(2\theta)$ e $\cos(2\theta)$')
plt.legend()
plt.grid(True)
plt.show()
```
